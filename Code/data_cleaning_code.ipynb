```python
import pandas as pd
import os

# Path to the main Excel file
excel_path = "C:/Users/alexr/OneDrive/Desktop/WORK/490_thesis/real_data_clean_one_file/490_data_main.xlsx"

# Path to the folder containing CSV files
csv_folder = "C:/Users/alexr/OneDrive/Desktop/WORK/490_thesis/5500_cleaned"

# Read the Excel file
excel_data = pd.read_excel(excel_path, sheet_name=None)

# Function to determine the CSV filename based on the year
def get_csv_filename(year):
    # Extract the year from the string (assuming format is 'Year_YYYY')
    year_num = int(year.split('_')[-1])
    if year_num < 2009:
        return f"f_5500_{year_num}_cleaned.csv"
    else:
        return f"f_5500_{year_num}_latest_cleaned.csv"

# Process each sheet and merge with corresponding CSV data
for year, df in excel_data.items():
    try:
        csv_filename = get_csv_filename(year)
        csv_path = os.path.join(csv_folder, csv_filename)
        
        if os.path.exists(csv_path):
            csv_data = pd.read_csv(csv_path)
            
            # Check for required columns
            for col in ['SPONS_DFE_MAIL_US_STATE', 'ADMIN_US_STATE']:
                if col not in csv_data.columns:
                    csv_data[col] = None  # Add column with None values if it doesn't exist
            
            # Select only the required columns from CSV data
            csv_data = csv_data[['ID', 'SPONS_DFE_MAIL_US_STATE', 'ADMIN_US_STATE']]
            
            # Merge with the Excel data
            merged_data = pd.merge(df, csv_data, on='ID', how='left')
            
            # Update the excel_data dictionary with the merged result
            excel_data[year] = merged_data
        else:
            print(f"CSV file for year {year} not found: {csv_filename}")
    except Exception as e:
        print(f"Error processing year {year}: {str(e)}")

# Create a new Excel writer object
output_path = "C:/Users/alexr/OneDrive/Desktop/WORK/490_thesis/real_data_clean_one_file/combined_data.xlsx"
with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
    # Write each dataframe to a separate sheet
    for year, df in excel_data.items():
        df.to_excel(writer, sheet_name=str(year), index=False)

print(f"Combined data saved to: {output_path}")
```

```python
import pandas as pd
import os

# Input and output file paths
input_file = "C:/Users/alexr/OneDrive/Desktop/WORK/490_thesis/safe_file/combined_data_1999-2023.xlsx"
output_folder = "C:/Users/alexr/OneDrive/Desktop/WORK/490_thesis/temp_files"
output_file = os.path.join(output_folder, "processed_combined_data_1999-2023.xlsx")

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Read the Excel file
xls = pd.ExcelFile(input_file)

# Process each sheet
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    for sheet_name in xls.sheet_names:
        # Read the sheet
        df = pd.read_excel(xls, sheet_name=sheet_name)
        
        # Convert columns to numeric, handling errors
        numeric_columns = ['INS_PRSN_COVERED_EOY_CNT', 'WLFR_PREMIUM_RCVD_AMT', 'WLFR_TOT_CHARGES_PAID_AMT']
        for col in numeric_columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        
        # Calculate new columns
        df['WLFR_PREMIUM_RCVD_AMT_PER_CAPITA'] = df.apply(
            lambda row: 0 if pd.isna(row['INS_PRSN_COVERED_EOY_CNT']) or row['INS_PRSN_COVERED_EOY_CNT'] == 0 
            else row['WLFR_PREMIUM_RCVD_AMT'] / row['INS_PRSN_COVERED_EOY_CNT'], 
            axis=1
        )
        
        df['WLFR_TOT_CHARGES_PAID_AMT_PER_CAPITA'] = df.apply(
            lambda row: 0 if pd.isna(row['INS_PRSN_COVERED_EOY_CNT']) or row['INS_PRSN_COVERED_EOY_CNT'] == 0 
            else row['WLFR_TOT_CHARGES_PAID_AMT'] / row['INS_PRSN_COVERED_EOY_CNT'], 
            axis=1
        )
        
        # Reorder columns to put new columns in G and H
        columns = df.columns.tolist()
        new_columns = [
            'WLFR_PREMIUM_RCVD_AMT_PER_CAPITA',
            'WLFR_TOT_CHARGES_PAID_AMT_PER_CAPITA'
        ]
        for col in new_columns:
            if col in columns:
                columns.remove(col)
        new_order = columns[:6] + new_columns + columns[6:]
        df = df.reindex(columns=new_order)
        
        # Write the processed dataframe to the new Excel file
        df.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"Processing complete. Output file saved as: {output_file}")
```
