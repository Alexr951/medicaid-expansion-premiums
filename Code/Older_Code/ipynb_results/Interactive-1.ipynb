```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Path to the file
path = r"C:/Users/alexr/OneDrive/Desktop/WORK/490_thesis/Final"

# Define ACA and non-ACA states
aca_states = ['AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'DC', 'HI', 'IL', 
              'IN', 'IA', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 
              'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 
              'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'UT', 'VT', 'VA', 'WA']

# Initialize lists to store data
all_data = []

# Read and process main data
excel_file = pd.ExcelFile(f"{path}/processed_combined_data_1999-2023.xlsx")

# Process each year's data
for year in range(1999, 2024):
    # Read sheet
    df = pd.read_excel(excel_file, sheet_name=f"Year_{year}")
    
    # Determine state column
    state_col = 'SPONS_DFE_MAIL_US_STATE' if year >= 2009 else 'SPONS_DFE_STATE'
    
    # Create ACA dummy
    df['ACA_State'] = df[state_col].isin(aca_states).astype(int)
    
    # Create post-treatment dummy
    df['Post_ACA'] = np.where(year >= 2010, 1, 0)
    
    # Create interaction term
    df['ACA_x_Post'] = df['ACA_State'] * df['Post_ACA']
    
    # Add year
    df['Year'] = year
    
    # Keep relevant columns
    cols_to_keep = ['Year', 'ACA_State', 'Post_ACA', 'ACA_x_Post',
                    'WLFR_PREMIUM_RCVD_AMT_PER_CAPITA',
                    'WLFR_TOT_CHARGES_PAID_AMT_PER_CAPITA']
    
    all_data.append(df[cols_to_keep])

# Combine all years
panel_data = pd.concat(all_data, ignore_index=True)

# Remove outliers using interquartile range
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Clean both dependent variables
panel_data = remove_outliers(panel_data, 'WLFR_PREMIUM_RCVD_AMT_PER_CAPITA')
panel_data = remove_outliers(panel_data, 'WLFR_TOT_CHARGES_PAID_AMT_PER_CAPITA')

# Run DiD regression for Premium Received
Y1 = panel_data['WLFR_PREMIUM_RCVD_AMT_PER_CAPITA']
X1 = panel_data[['ACA_State', 'Post_ACA', 'ACA_x_Post']]
X1 = sm.add_constant(X1)

model1 = sm.OLS(Y1, X1)
results1 = model1.fit()

# Run DiD regression for Total Charges Paid
Y2 = panel_data['WLFR_TOT_CHARGES_PAID_AMT_PER_CAPITA']
X2 = X1  # Same independent variables

model2 = sm.OLS(Y2, X2)
results2 = model2.fit()

# Create summary DataFrames
def create_results_df(results):
    return pd.DataFrame({
        'Coefficient': results.params,
        'Std Error': results.bse,
        't-stat': results.tvalues,
        'p-value': results.pvalues,
        'CI Lower': results.conf_int()[0],
        'CI Upper': results.conf_int()[1]
    })

# Create results DataFrames
results_df1 = create_results_df(results1)
results_df2 = create_results_df(results2)

# Print basic results
print("\nResults for Premium Received (Experience-Rated Plans):")
print(results_df1)
print("\nR-squared:", results1.rsquared)
print("\nResults for Total Charges Paid (Non-Experience-Rated Plans):")
print(results_df2)
print("\nR-squared:", results2.rsquared)

# Save results to Excel
with pd.ExcelWriter(f"{path}/regression_results_cleaned.xlsx") as writer:
    results_df1.to_excel(writer, sheet_name='Premium_Received_Results')
    results_df2.to_excel(writer, sheet_name='Total_Charges_Results')
```
